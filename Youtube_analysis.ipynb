```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Load Dataset

Dataset taken from Kaggle.
https://www.kaggle.com/datasets/mayankanand2701/top-1000-youtube-channels-in-the-world/data

```python
df = pd.read_csv("/content/drive/MyDrive/Youtubers Analysis/Top Youtubers Dataset.csv", encoding= 'latin1')
```

Definition of each columns

*   Rank : The ranking of the YouTube channel based on its overall popularity and performance.
*   Youtuber : The name of the YouTuber or the title of the YouTube channel.
*   Subscribers : The total number of subscribers to the channel, indicating its reach and popularity.
*   Video Views : The total number of video views the channel has accumulated, reflecting its engagement and audience interaction.
*   Video Count : The total number of videos uploaded by the channel, showing the content volume produced.
*   Category : The genre or category the channel belongs to, such as music, education, entertainment, etc.
*   Started : The year the channel was created, providing insight into its longevity and growth over time.

# Visualization

```python
df.Category.value_counts().plot(kind='bar')
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Subscribers'], bins=100, kde=True)
plt.title('Subscribers Distribution')
plt.xlabel('Subscribers')
plt.ylabel('Frequency')
plt.show()
```

```python
top_channels = df.sort_values(by='Subscribers', ascending=False).head(10)
print(top_channels)
```

```python
plt.figure(figsize=(10, 5))
sns.barplot(x=top_channels['Subscribers'], y=top_channels['Youtuber'])
plt.title('Top 10 Channels by Subscribers')
plt.show()
```

```python
top_views = df.sort_values(by='Video Views', ascending=False).head(10)
print(top_views)
```

```python
plt.figure(figsize=(10, 5))
sns.barplot(x=top_views['Video Views'], y=top_views['Youtuber'])
plt.title('Top 10 Channels by Video Views')
plt.show()
```

# Entertainment Analysis

```python
Entertainment['Started'].value_counts()
```

```python
year_started = Entertainment['Started'].value_counts().to_dict()
sorted_dict = {str(k): year_started[k] for k in sorted(year_started)}

plt.figure(figsize = (7, 5))
sns.lineplot(x = sorted_dict.keys(), y = sorted_dict.values())
plt.xticks(rotation = 45)
plt.xlabel('Year')
plt.ylabel('No of Entertainment YouTubers Starting')
plt.title('Number of Entertainment YouTubers joining across different years')
```

*   YouTubers with entertainment categories have the largest number, while YouTubers with autos & vehicels categories have the smallest number.
